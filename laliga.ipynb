```python
import requests
import re
import json
url = "https://en.wikipedia.org/wiki/List_of_Spanish_football_champions"
page = requests.get(url).text
#print(page)
```

```python
tables = re.findall(r"\<tbody>\<tr\>(.*?)\</tbody\>\</table\>", page, re.DOTALL)
```

```python
main_table = tables[2]
headers = re.findall(r"\<th(.*?)th\>", main_table, re.DOTALL)
print(main_table)
#print(headers)
```

```python
breakdown = re.findall(r"\<tr\>(.*?)\</tr\>", main_table, re.DOTALL)

seasons_dict = {}
for i in breakdown:
    try:
        cleaned_row_list=[]
        columns = re.findall(r"\<td(.*?)\</td\>", i, re.DOTALL)
        #print(columns)
        for j in columns:
            clean_j= re.sub(r"\<(.*?)\>", "", j, re.DOTALL)
            clean_j = "".join(re.sub(r"(.*?)\>","", clean_j, re.DOTALL).split("\n"))
            #print(clean_j)
            cleaned_row_list.append(clean_j)
        #print(cleaned_row_list)
        seasons_dict[cleaned_row_list[0][-2:]] = {"1st Place":cleaned_row_list[1],"2nd Place":cleaned_row_list[2], "3rd Place":cleaned_row_list[3],"Top Scorer":cleaned_row_list[4], "Top Scorer's Club":cleaned_row_list[5], "Goals":cleaned_row_list[6]}
    except:
        seasons_dict[cleaned_row_list[0][-2:]] = "Spanish Civil War (League Cancelled)"
        continue
#print(seasons_dict["1929"])
print(json.dumps(seasons_dict, indent=3))
```
